# Initialisation

```python
from google.colab import files
import sqlite3
import pandas as pd

!rm Database.db

bdd = sqlite3.connect("Database.db")
curseur = bdd.cursor()

def afficher(query):
    visualisation = pd.read_sql_query(query, bdd)
    return visualisation
```

# Création de la Base de Données

Table Joka

```python
tableJoka = """
CREATE TABLE Joka (
    ID_Joka INTEGER PRIMARY KEY,
    Nom STRING,
    Vie INTEGER
)
"""

curseur.execute(tableJoka)

curseur.execute("""
INSERT INTO Joka
VALUES

    (0, "Glimmergeist", 100),
    (1, "Duskwing", 100),
    (2, "Thunderhoof", 100),
    (3, "Ironsoul", 100),
    (4, "Mysticgaze", 100),
    (5, "CyberZap", 100),
    (6, "Infernite", 100),
    (7, "Luminex", 100),
    (8, "Toxiclone", 100),
    (9, "Electrogolem", 100),
    (10, "Chronoblitz", 100),
    (11, "Entropicore", 100),
    (12, "Hailstormer", 100),
    (13, "Cryoquake", 100),
    (14, "Spectralgeist", 100),
    (15, "Phantom Prowler", 100),
    (16, "Shadow Mermaid", 100),
    (17, "Blaze Brute", 100),
    (18, "Silver Queen", 100),
    (19, "Solar Serpent", 100),
    (20, "Frosty Behemoth", 100)

""")

bdd.commit()
afficher("SELECT * FROM Joka")
```

Table Inventaire

```python
tableInventaire = """
CREATE TABLE Inventaire (
    ID_Joka INTEGER PRIMARY KEY,
    estDansInventaire STRING,
    FOREIGN KEY (ID_Joka) REFERENCES Joka(ID_Joka)
)
"""

curseur.execute(tableInventaire)

bdd.commit()
```

Table Graphe

```python
tableGraphe = """
CREATE TABLE Graphe (
    PositionX INTEGER,
    PositionY INTEGER,
    Nom STRING,
    Voisins STRING,
    PRIMARY KEY (PositionX, PositionY)
)
"""

curseur.execute(tableGraphe)

curseur.execute("""
INSERT INTO Graphe
VALUES

    (0, 0, 'A', '["B", "C"]'),
    (1, 0, 'B', '["A"]'),
    (1, 1, 'C', '["A", "D"]'),
    (0, 2, 'D', '["C"]')

""")

bdd.commit()
afficher("""
    SELECT * FROM Graphe
""")
```

Table Technique

```python
tableTechnique = """
CREATE TABLE Technique (
    ID_Technique INTEGER PRIMARY KEY,
    Nom STRING,
    Type STRING,
    Puissance INTEGER
)
"""

curseur.execute(tableTechnique)

curseur.execute("""
INSERT INTO Technique
VALUES

    (0, "Attaque 1", "Attaque", 10),
    (1, "Attaque 2", "Attaque", 5),
    (2, "Attaque 3", "Attaque", 15),
    (3, "Attaque 4", "Attaque", 5),
    (4, "Défense 1", "Défense", 10),
    (5, "Défense 2", "Défense", 5),
    (6, "Défense 3", "Défense", 15),
    (7, "Défense 4", "Défense", 5)

""")

bdd.commit()
afficher("SELECT * FROM Technique")
```

Table Association

```python
tableAssociation = """
CREATE TABLE Association (
    ID_Joka INTEGER,
    ID_Technique INTEGER,
    FOREIGN KEY (ID_Joka) REFERENCES Joka(ID_Joka)
    FOREIGN KEY (ID_Technique) REFERENCES Technique(ID_Technique)
)
"""

curseur.execute(tableAssociation)

curseur.execute("""
INSERT INTO Association
VALUES

    (0, 0),
    (0, 1),
    (0, 4),
    (1, 2),
    (1, 3),
    (1, 5)

""")

bdd.commit()
afficher("""
    SELECT Association.ID_Joka, Joka.Nom, Association.ID_Technique, Technique.Nom
    FROM Association
    JOIN Joka ON Joka.ID_Joka = Association.ID_Joka
    JOIN Technique ON Technique.ID_Technique = Association.ID_Technique
""")
```

Table Item

```python
tableItem = """
CREATE TABLE Item (
    ID_Item INTEGER PRIMARY KEY,
    Nom STRING,
    Type STRING,
    Puissance INTEGER
)
"""

curseur.execute(tableItem)

curseur.execute("""
INSERT INTO Item
VALUES

    (0, "Puissance 1", "Puissance", 10),
    (1, "Puissance 2", "Puissance", 5),
    (2, "Puissance 3", "Puissance", 20),
    (3, "Potion 1", "Soin", 5),
    (4, "Potion 2", "Soin", 25),
    (5, "Potion 3", "Soin", 20)

""")

bdd.commit()
afficher("SELECT * FROM Item")
```
